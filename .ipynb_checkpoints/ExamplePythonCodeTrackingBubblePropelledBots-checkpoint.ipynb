```python
def threshold(image, sigma=3):
    mask = image > (image.mean() + sigma*image.std())#sigma should be postive for an inverted image
    return mask 
```

```python
def Batch_BF_BubbleTrack(vid,BackgroundImg,firstimg,lastimg,sz_thresh=50):
    data = []
    for j in np.linspace(firstimg,lastimg,lastimg-firstimg+1,dtype='int'):
        frame_number = j
        vid.set(1, frame_number-1)
        suc,im1 = vid.read()
        im1 = cv.cvtColor(im1,cv.COLOR_BGR2GRAY);

        im1blur = im1-BackgroundImg #this line and the next are additions to the florescent version
        im1blur = im1blur-np.min(im1blur)
        #im1blur=scipy.ndimage.gaussian_filter(im1blur,1)
        im1blur = np.uint8(im1blur)

        mask = threshold(255-im1blur, sigma = 3)#brightfield so invert
        label_im, nb_labels = ndimage.label(mask) 
        sizes = ndimage.sum(mask, label_im, range(nb_labels + 1)) 

        data_all = []
        num_bots=np.sum(sizes>sz_thresh)
        if  num_bots>0:
            szsorted=np.argsort(sizes)
            for i in range(num_bots):
                [ycord,xcord]=ndimage.center_of_mass(im1blur,labels=label_im,index = szsorted[-(i+1)])
                data1 = [frame_number,xcord,ycord,sizes[szsorted[-(i+1)]]]
                data_all = np.concatenate((data_all,data1),axis=0)
            data = np.concatenate((data,data_all),axis=0)
    data = np.reshape(data,[int(len(data)/4),4])
    return data
```

```python
def Bubble_Bright_FluorPart2V2(vid,BackgroundImg,BF,indicesBF,firstframe=1,sz_thresh=50):
    data = []
    suc,im1 = vid.read()#read the first image
    im1 = cv.cvtColor(im12,cv.COLOR_BGR2GRAY);
    dim=np.shape(im1)

    #indx, indy = np.meshgrid(range(dim[1]),range(dim[0]))#the indices in matrix form

    #fps = vid.get(cv.CAP_PROP_FPS)
    totalNoFrames = vid.get(cv.CAP_PROP_FRAME_COUNT);
    totalNoFrames=int(totalNoFrames)
    region = 0
    sz_threshF = 10 #min size of fluorescent particles

    #for i in range(firstframe,totalNoFrames):
    for i in range(firstframe,2164):
        frame_number = i
        vid.set(1, frame_number-1)
        suc,im1 = vid.read()

        # if the current frame is in the next region, increase region by 1
        if i >= indicesBF[region+1]:
            region = region+1
        if BF[region] == 0: #if it is fluorescent, no need to subtract background
            im1[:,:,2]=0#eliminate the red channel since that is the color of the scalebar and time
            im1 = cv.cvtColor(im1,cv.COLOR_BGR2GRAY);
            im1blur = im1.astype('float')
            im1blur = np.uint8(im1blur)
            mask = threshold(im1blur, sigma = 10)
            label_im, nb_labels = ndimage.label(mask) 
            sizes = ndimage.sum(mask, label_im, range(nb_labels + 1)) 
            data_all = []
            num_bots=np.sum(sizes>sz_threshF)
            if  num_bots>0:
                szsorted=np.argsort(sizes)
                for i in range(num_bots):
                    [ycord,xcord]=ndimage.center_of_mass(im1blur,labels=label_im,index = szsorted[-(i+1)])
                    data1 = [frame_number,xcord,ycord,sizes[szsorted[-(i+1)]],0]
                    data_all = np.concatenate((data_all,data1),axis=0)
                data = np.concatenate((data,data_all),axis=0)
        else:#if it is brifhtfield
            im1 = cv.cvtColor(im1,cv.COLOR_BGR2GRAY);
            im1blur = im1.astype('float')
            im1blur[im1blur>255] = 255
            im1blur[im1blur<0]=0
            im1blur = im1blur-BackgroundImg #this line and the next are additions to the florescent version
            im1blur = im1blur-np.min(im1blur)
            #im1blur=scipy.ndimage.gaussian_filter(im1blur,1)
            im1blur = np.uint8(im1blur)

            mask = threshold(255-im1blur, sigma = 10)#brightfield so invert
            label_im, nb_labels = ndimage.label(mask) 
            sizes = ndimage.sum(mask, label_im, range(nb_labels + 1)) 

            data_all = []
            num_bots=np.sum(sizes>sz_thresh)
            if  num_bots>0:
                szsorted=np.argsort(sizes)
                for i in range(num_bots):
                    [ycord,xcord]=ndimage.center_of_mass(im1blur,labels=label_im,index = szsorted[-(i+1)])
                    data1 = [frame_number,xcord,ycord,sizes[szsorted[-(i+1)]],1]
                    data_all = np.concatenate((data_all,data1),axis=0)
                data = np.concatenate((data,data_all),axis=0)
    data = np.reshape(data,[int(len(data)/5),5])
    return data
```

### Determining Bubble frequency:

```python
bubblefreqs = np.zeros([len(particle_nums),np.sum(BF)+1])*np.nan
for i in range(len(particle_nums)):
    for j in range(np.sum(BF)):
        particle_data1 = t1[t1.particle == particle_nums[i]]
        #particle_data1 = particle_data1[particle_data1.frame>indicesBF[1]]
        particle_data1 = particle_data1[np.all([particle_data1.frame>indicesBF[np.abs(BF[0]-1)+j*2],particle_data1.frame<indicesBF[np.abs(BF[0]-1)+1+j*2]],axis=0)]
        #particle_data1.plot(x='frame',y='sz')
        if len(particle_data1) > 20:
            new_index = pd.Index(np.arange(np.min(particle_data1.frame),np.max(particle_data1.frame),1), name="frame")
            particle_data1=particle_data1.set_index("frame").reindex(new_index)
            bubblesz = np.array(particle_data1.sz)

            #interpolate nans
            mask = np.isfinite(bubblesz)
            xi = np.arange(len(bubblesz))
            bubblesz = np.interp(xi, xi[mask], bubblesz[mask])

            bubblefft = rfft(bubblesz)
            freqs = rfftfreq(len(bubblesz), 1)[:len(bubblesz)//2]
            freqs=freqs[1:]
            fftdata=2.0/len(bubblesz) * np.abs(bubblefft[0:len(bubblesz)//2])[1:]
            if fftdata[np.argmax(fftdata)] > np.mean(fftdata)+2*np.std(fftdata):
                bubblefreqs[i,j]=freqs[np.argmax(fftdata)]
                bubblefreqs[i,-1]=particle_nums[i]
```

### Determine min bubble size:

```python
minBubblesz = np.zeros([len(particle_nums),np.sum(BF)+1])*np.nan
minBubblesz_num = np.zeros([len(particle_nums),np.sum(BF)+1])*np.nan
for i in range(len(particle_nums)):
    for j in range(np.sum(BF)):
        particle_data1 = t1[t1.particle == particle_nums[i]]
        particle_data1 = particle_data1[np.all([particle_data1.frame>indicesBF[np.abs(BF[0]-1)+j*2],particle_data1.frame<indicesBF[np.abs(BF[0]-1)+1+j*2]],axis=0)]
        if len(particle_data1) > 20:
            inc = 10 #find the max bubble size in every inc frames. This assumes that the period is equal or less than this, so confirm this before running or increase if it isn't
            num = np.int(np.ceil(len(particle_data1)/inc))
            minvals = np.zeros(num)*np.nan
            for ii in range(num):
                minvals[ii] = np.min(particle_data1[(ii)*inc:(ii+1)*inc].sz)
            minBubblesz[i,j]=np.nanmedian(minvals)
            minBubblesz_num[i,j] = len(particle_data1)
            minBubblesz[i,-1] = particle_nums[i]
            minBubblesz_num[i,-1] = particle_nums[i]
```

### Determine maximum bubble size:

```python
maxBubblesz = np.zeros([len(particle_nums),np.sum(BF)+1])*np.nan
maxBubblesz_num = np.zeros([len(particle_nums),np.sum(BF)+1])*np.nan
for i in range(len(particle_nums)):
    for j in range(np.sum(BF)):
        particle_data1 = t1[t1.particle == particle_nums[i]]
        particle_data1 = particle_data1[np.all([particle_data1.frame>indicesBF[np.abs(BF[0]-1)+j*2],particle_data1.frame<indicesBF[np.abs(BF[0]-1)+1+j*2]],axis=0)]
        if len(particle_data1) > 20:
            inc = 10 #find the max bubble size in every inc frames. This assumes that the period is equal or less than this, so confirm this before running or increase if it isn't
            num = np.int(np.ceil(len(particle_data1)/inc))
            maxvals = np.zeros(num)*np.nan
            for ii in range(num):
                maxvals[ii] = np.max(particle_data1[(ii)*inc:(ii+1)*inc].sz)
            maxBubblesz[i,j]=np.nanmedian(maxvals)
            maxBubblesz_num[i,j] = len(particle_data1)
            maxBubblesz[i,-1] = particle_nums[i]
            maxBubblesz_num[i,-1] = particle_nums[i]
```

### Determine the average particle speed

```python
speeds = np.nan*np.zeros([len(particle_nums),len(BF)+1])#each row corresponds to each particle and each column is each Fluor/BF region, last col is particle num
speeds_nums = np.nan*np.zeros([len(particle_nums),len(BF)+1])
for i in range(len(particle_nums)):
    for j in range(len(BF)):
        particle_data1 = t1[t1.particle == particle_nums[i]]
        particle_data1 = particle_data1[np.all([particle_data1.frame>indicesBF[j],particle_data1.frame<indicesBF[j+1]],axis=0)]
        if len(particle_data1.x) > 20:
            particle_data1.set_index("frame")#set missing data to nan
            new_index = pd.Index(np.arange(np.min(particle_data1.frame),np.max(particle_data1.frame),1), name="frame")
            particle_data1=particle_data1.set_index("frame").reindex(new_index)
            particle_data1_smoothed=particle_data1.rolling(20,center=True,min_periods=5).mean()#smooth the data
            particle_data1_smoothed = particle_data1_smoothed[particle_data1_smoothed.particle == particle_nums[i]]
            #fit to polynomials:
            xsmoothed = UnivariateSpline(np.array(particle_data1_smoothed.index,dtype='int'), np.array(particle_data1_smoothed.x))
            ysmoothed = UnivariateSpline(np.array(particle_data1_smoothed.index,dtype='int'), np.array(particle_data1_smoothed.y))
            xsmoothed.set_smoothing_factor(20)
            ysmoothed.set_smoothing_factor(20)#smooth the fit
            vx = xsmoothed.derivative()
            vy = ysmoothed.derivative()#find the velocity
            speed = np.sqrt(vy(particle_data1_smoothed.index)**2+vx(particle_data1_smoothed.index)**2)#find the speed
            speeds[i,j] = np.mean(speed[5:-5])#find the average speed, ignore edges bc the fits prob aren't great there
            speeds[i,-1] = particle_nums[i]
            speeds_nums[i,j] = len(particle_data1)-20
            speeds_nums[i,-1] = particle_nums[i]
```
